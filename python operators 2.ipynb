Assignment operators and comparison operators.

Assignment Operators :- Assignment operators are used to assign values to variables.

```python
#  Operator  Example    Same As
#  =         x = 5      x = 5
#  +=        x += 3     x = x + 3
#  -=        x -= 3     x = x - 3
#  *=        x *= 3     x = x * 3
#  /=        x /= 3     x = x / 3
#  %=        x %= 3     x = x % 3
#  //=       x //= 3    x = x // 3
#  **=       x **= 3    x = x ** 3
#  &=        x &= 3     x = x & 3
#  |=        x |= 3     x = x | 3
#  ^=        x ^= 3     x = x ^ 3
#  >>=       x >>= 3    x = x >> 3
#  <<=       x <<= 3    x = x << 3
```

```python
# = Operator :- 

x = 5
print("x= ", x)
```

```python
# += Operator:-

x += 5
print(x)
```

```python
# -= Operator :-

x -= 5
print(x)
```

```python
# *= Operator :-

x *= 5
print(x)
```

```python
# /= Operator :-

x /= 5
print(x)
```

```python
# %= Operator

x = 18
x %= 5
print(x)
```

```python
# //= Operator
x = 5
x //= 2
print(x)
```

```python
# **= Operator

x **= 5
print(x)
```

```python
# &= Operator
x = 5
x &= 3
print(x)
```

```python
# |= Operator

x = 10
x |= 3
print(x)
```

```python
# ^= Operator

x ^= 2
print(x)
```

```python
# >>= Operator

x >>= 2
print(x)
```

```python
# <<= Operator

x = 10
x <<= 2
print(x)
```

Comparison Operators :- Comparison operators are used to compare two values.

```python
#  Operator  Name                     Example
#  ==       Equal                      x == y
#  !=       Not equal                  x != y
#  >        Greater than               x > y
#  <        Less than                  x < y
#  >=       Greater than or equal to   x >= y
#  <=       Less than or equal to      x <= y 
```

```python
# Equal Operator (==) :- It compares the equality of two variables or values.

x = 5
y = 10
print(x == y)
```

```python
# Not Equal Operator (!=) :- It checks the non-equality of two variables or values. It is opposite of the equal operator.

x = 5
y = 10
print(x != y)
```

```python
# Greater Than Operator (>) :- It checks whether a given variable or value is greater than the second.

x = 5
y = 10
print(x > y)
```

```python
# Less Than Operator (<) :- It checks whether a given variable or value is less than the second.

x = 5
y = 10
print(x < y)
```

```python
# Greater Than or Equal To Operator (>=) :- It checks if a variable or value is either greater than or equal to the other variable or value. If any of the two condition is right.
# It returns true otherwise false.

x = 5
y = 10
print(x >= y)
```

```python
# Less Than or Equal To Operator (<=) :- It checks if a variable or value is either less than or equal to the other variable or value. If any of the two condition is right.
# It returns true otherwise false.

x = 5
y = 10
print(x <= y)
```
